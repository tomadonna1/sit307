```python
import pandas as pd           
import numpy as np          
import matplotlib.pyplot as plt
import seaborn as sns    
import scipy                  
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import QuantileTransformer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
from cuml import PCA as cumlPCA  
import cupy as cp
from mpl_toolkits.mplot3d import Axes3D
from cuml.model_selection import GridSearchCV
from cuml.svm import SVC
from cuml.neighbors import KNeighborsClassifier
```

# Preprocess data

```python
df = pd.read_csv('./8.1P-resources/Dataset3.csv', header=None)
df = df[0].str.split(';', expand=True)
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)
df
```

```python
df.info()
```

```python
df['Target'].value_counts()
```

```python
# Label encode Target
le = LabelEncoder()
df['Target'] = le.fit_transform(df['Target'])
df['Target'].value_counts()
```

There seems to be imbalance between classes, I will oversample the undersample class.

```python
X = df.drop('Target', axis=1)
y = df.Target
```

```python
# Apply oversampling
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Convert resampled data back into a pandas DataFrame
X_resampled = pd.DataFrame(X_resampled, columns=X.columns)
y_resampled = pd.DataFrame(y_resampled, columns=['Target'])
```

```python
y_resampled.value_counts()
```

Now we need to quantile transform our X_resampled.

```python
# Convert the X_samppled numeric and fill the nan with mean 
X_resampled = X_resampled.apply(pd.to_numeric, errors='coerce')
X_resampled = X_resampled.fillna(X_resampled.mean())
X_resampled.head()
```

```python
# Check the unique value of each feature
unique_values = X_resampled.nunique()
categorical_columns = unique_values[unique_values < 10].index.tolist()
print("Categorical columns based on unique values:", categorical_columns)
```

```python
# Separate numerical columns (excluding the specified columns)
numerical_columns = [col for col in X_resampled.columns if col not in categorical_columns]

# Apply QuantileTransformer only to the numerical columns
qt = QuantileTransformer(output_distribution='normal', random_state=42)
X_resampled_numerical_transformed = qt.fit_transform(X_resampled[numerical_columns])

# Convert the transformed numerical data back to a DataFrame
X_resampled_numerical_transformed_df = pd.DataFrame(X_resampled_numerical_transformed, columns=numerical_columns)

# Combine the transformed numerical data with the excluded categorical data
X_resampled_transformed_df = pd.concat([X_resampled[categorical_columns].reset_index(drop=True), 
                                        X_resampled_numerical_transformed_df.reset_index(drop=True)], axis=1)

# Display the resulting transformed dataset
X_resampled_transformed_df.head()
```

# 1. Apply PCA

```python
# Perform PCA
pca = cumlPCA(n_components=None)
X_pca = pca.fit(X_resampled_transformed_df)

# Explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
display(cumulative_variance)

# Plot the cumulative_variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance')
plt.grid(True)
plt.show()
```

18 PC seems to be the sweet spot (explains 95% of variance). We can visualize with the first 3 PC. 

```python
# Perform PCA
pca = cumlPCA(n_components=18)
X_pca = pca.fit_transform(X_resampled_transformed_df)
X_pca.head()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[0], X_pca[1], X_pca[2], c=y_resampled, cmap='viridis', s=50, alpha=0.7)
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
plt.title('3D Visualization of First Three Principal Components')
plt.show()
```

It is quite hard to visualize 3D, and also 3 PC only explains 53% of the variance.

# 2. Fit SVM model
I will use GridSearch to find the best combination of hyperparameters.

```python
C_values = np.logspace(-3, 2, 10)
C_values_normal = [f'{c:.4f}' for c in C_values]
C_values_normal
```

```python
# Split the data
y_resampled_array = np.array(y_resampled)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_resampled_array.ravel(), test_size=0.3, random_state=42)

# SVC model
svm = SVC()

# Grid search
param_grid = {
    'C': np.logspace(-3, 2, 10),
    'kernel': ['rbf'],
    'gamma': ['scale', 'auto']
}
grid_search = GridSearchCV(svm, param_grid, cv=5)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
print(f"Best hyperparameters: {best_params}")
```

```python
# Train the best SVM model
best_svm = grid_search.best_estimator_
best_svm.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = best_svm.score(X_test, y_test)
print(f"Test set accuracy: {accuracy}")
```

Seems like we need a high `C` parameter

```python
C_values = np.linspace(10, 40, 40)
C_values = [f'{c:.4f}' for c in C_values]
C_values
```

```python
# Split the data
y_resampled_array = np.array(y_resampled)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_resampled_array.ravel(), test_size=0.3, random_state=42)

# SVC model
svm = SVC()

# Grid search
param_grid = {
    'C': np.linspace(10, 40, 40),
    'kernel': ['rbf'],
    'gamma': ['scale', 'auto']
}
grid_search = GridSearchCV(svm, param_grid, cv=5)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
print(f"Best hyperparameters: {best_params}")

# Train the best SVM model
best_svm = grid_search.best_estimator_
best_svm.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = best_svm.score(X_test, y_test)
print(f"Test set accuracy: {accuracy}")
```

Best hyperparameters: {'C': 19.230769230769234, 'gamma': 'auto', 'kernel': 'rbf'}

Test set accuracy: 0.798893928527832

Seems like the value of around 19, 20 is the optimal `C`. I will test more metrics

```python
# Predict the labels
y_pred = best_svm.predict(X_test)

# classification report
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
y_pred_prob = best_svm.decision_function(X_test)
# Compute ROC curve and ROC AUC score for each class
fpr = {}
tpr = {}
roc_auc = {}

for i in range(len(np.unique(y_test))):  # Loop over each class
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_pred_prob[i])
    roc_auc[i] = roc_auc_score(y_test == i, y_pred_prob[i])

# Plot ROC curve for each class
plt.figure(figsize=(10, 8))
for i in range(len(np.unique(y_test))):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for SVM')
plt.legend(loc='lower right')
plt.show()
```

Here is my take on Question 2. So for the data splitting, its nothing much, just a regular 70/30 split. The data used to split is the resampled and scaled data from the preprocessing step. For the metrics, I have used classification report and roc curve. The classification report shows a report on 4 metrics, which are precision, recall, f1 and support. The report shows balance performance on all dataset, though slightly underperforms on class 2. The overall accuracy is 80% on 1989 samples. The roc curve shows how well our classifer distinguish between classes, and based on the number I think the classifer did a decent job.

# 4 and 5. KNN using varying distances and hyperparameter

```python
# Grid search
knn = KNeighborsClassifier()
param_grid_knn = {
    'n_neighbors': list(range(1, 101)),         
    'weights': ['uniform', 'distance'],  
    'metric': ['euclidean', 'manhattan', 'minkowski'],
    'p': [1, 2, 3], # Power parameter for the Minkowski metric                                            
}
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=5, scoring='accuracy')
grid_search_knn.fit(X_train, y_train)
best_params_knn = grid_search_knn.best_params_
print(f"Best hyperparameters for KNN: {best_params_knn}")

# Evaluate the best model on the test set
best_knn = grid_search_knn.best_estimator_
accuracy_knn = best_knn.score(X_test, y_test)
print(f"Test set accuracy: {accuracy_knn}")
```

Best hyperparameters for KNN: {'metric': 'euclidean', 'n_neighbors': 1, 'p': 1, 'weights': 'uniform'}

Test set accuracy: 0.8034188151359558

```python
# Predict the labels
y_pred = best_knn.predict(X_test)

# classification report
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
y_proba = best_knn.predict_proba(X_test)
n_classes = y_proba.shape[1]
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test == i, y_proba[i])
    roc_auc = roc_auc_score(y_test == i, y_proba[i])
    
    # Plot ROC curve for each class
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Multi-Class KNN')
plt.legend(loc='lower right')
plt.show()
```

I have used the same metrics as SVM on KNN. For the classification report, the two algorithms did pretty much the same. On the ROC curve, SVM did better. For the tuning of the hyperparameter, I have chosen a few distance metrics to find the most optimal measurement for our data. I can't really visualize so I ought to test a couple of them.
